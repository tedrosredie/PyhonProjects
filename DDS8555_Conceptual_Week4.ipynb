```python
import numpy as np
import matplotlib.pyplot as plt

# Define the basis functions
b1 = lambda x: x
b2 = lambda x: (x - 1)**2 * (x >= 1)  # (X - 1)^2 if X >= 1 else 0

# Coefficients
beta0 = 1
beta1 = 1
beta2 = -2

# Define the function Y
Y = lambda x: beta0 + beta1 * b1(x) + beta2 * b2(x)

# Generate X values
X = np.linspace(-2, 2, 400)

# Calculate Y values
Y_vals = Y(X)

# Plotting
plt.plot(X, Y_vals, label='Estimated curve', color='c')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Estimated Curve from X = -2 to X = 2')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.legend()
plt.grid(True)
plt.show()
```
