Data Loading and Summary

```python
import pandas as pd

df = pd.read_csv('/content/baby.csv')
print(df.head())
print(df.info())
print(df.describe(include='all'))
```

Data Cleaning

```python
print(df.isnull().sum())
# If missing values exist, display rows with NaNs
print(df[df.isnull().any(axis=1)])
# Handle missing data (example: drop or fill)
# df = df.dropna()
```

Univariate Analysis

```python
import matplotlib.pyplot as plt
import seaborn as sns

num_cols = ['Birth Weight', 'Gestational Days', 'Maternal Age', 'Maternal Height', 'Maternal Pregnancy Weight']
for col in num_cols:
    plt.figure(figsize=(6,4))
    sns.histplot(df[col], kde=True)
    plt.title(f'Distribution of {col}')
    plt.show()
```

Maternal Smoking Analysis

```python
print(df['Maternal Smoker'].value_counts())
sns.countplot(data=df, x='Maternal Smoker')
plt.title('Count of Maternal Smoking Status')
plt.show()
```

Bivariate Relationships

```python
# Boxplot: Birth Weight by Maternal Smoker
sns.boxplot(x='Maternal Smoker', y='Birth Weight', data=df)
plt.title('Birth Weight by Maternal Smoking Status')
plt.show()

# Scatterplot: Birth Weight vs Gestational Days
sns.scatterplot(x='Gestational Days', y='Birth Weight', hue='Maternal Smoker', data=df)
plt.title('Birth Weight vs Gestational Days')
plt.show()
```

Correlation Analysis

```python
corr = df[num_cols].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

# **Key Insights**

Birth Weight appears affected by gestational period and maternal smoking.

Maternal Smoking is associated with lower average birth weight (as shown by the boxplot).

Several predictors (gestational days, maternal weight) are positively correlated with birth weight.
