```python
import numpy as np

x = np.arange(1, 21, 1)
y = np.array([1, 3, 5, 7, 9, 12, 15, 19, 23, 28,
              33, 38, 44, 50, 56, 64, 73, 84, 97, 113])
```

```python
import matplotlib.pyplot as plt
plt.scatter(x,y)
plt.show()
```

```python
#fit the model
fit = np.polyfit(x, np.log(y),1)
```

```python
#view the output of the model
print(fit)
```
