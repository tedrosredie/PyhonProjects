**K-Means Clustering Example**

A step-by-step implementation of K-Means clustering algorithm using Python in Google Colab. This project demonstrates the complete machine learning workflow for unsupervised learning, from data generation to evaluation.

**Features**

* Synthetic dataset generation with customizable clusters

* Elbow method for determining optimal cluster count

* K-Means clustering implementation

* Visualization of clusters and centroids

* Performance evaluation using silhouette score

* Prediction on new data points

**Technologies Used**

* Python 3

* NumPy for numerical operations

* Matplotlib for visualization

* Scikit-learn for machine learning algorithms

Cluster-2.svg

**Step 1: Install & Import Libraries**

```python
# No installations needed in Colab for these libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

**Step 2: Generate Sample Data**

```python
# Create synthetic dataset with 4 clusters
X, y_true = make_blobs(
    n_samples=300,
    centers=4,
    cluster_std=0.60,
    random_state=42
)

plt.scatter(X[:, 0], X[:, 1], s=50)
plt.title("Raw Data")
plt.show()
```

**Step 3: Determine Optimal Number of Clusters**

```python
# Elbow Method
wcss = []  # Within-Cluster-Sum-of-Squares
k_range = range(1, 11)

for k in k_range:
    kmeans = KMeans(n_clusters=k, n_init=10, random_state=42)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)

# Plot elbow curve
plt.plot(k_range, wcss, 'bx-')
plt.xlabel('Number of clusters (k)')
plt.ylabel('WCSS')
plt.title('Elbow Method')
plt.show()
```

**Step 4: Apply K-Means Clustering**

```python
# Based on elbow plot, choose k=4
kmeans = KMeans(n_clusters=4, n_init=10, random_state=42)
y_pred = kmeans.fit_predict(X)

# Get cluster centers
centroids = kmeans.cluster_centers_
```

**Step 5: Visualize Clusters**

```python
plt.scatter(X[:, 0], X[:, 1], c=y_pred, s=50, cmap='viridis')
plt.scatter(centroids[:, 0], centroids[:, 1], c='red', s=200, alpha=0.75, marker='X')
plt.title('K-Means Clustering Results')
plt.show()
```

**Step 6: Evaluate Clustering (Optional)**

```python
# Silhouette Score (-1 to 1, higher is better)
score = silhouette_score(X, y_pred)
print(f"Silhouette Score: {score:.3f}")
```

**Step 7: Make Predictions on New Data**

```python
# Predict cluster for new samples
new_points = np.array([[-2, -3], [-5, -5], [3, 4]])
new_preds = kmeans.predict(new_points)

print("Predicted clusters for new points:")
for point, cluster in zip(new_points, new_preds):
    print(f"Point {point} -> Cluster {cluster}")
```

**Results**

The algorithm successfully identifies clusters in synthetic data, visualizes the results with centroid markers, and provides quantitative evaluation of clustering quality.
